```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #need to install seaborn "conda install seaborn"
```

```python
df = pd.read_csv('test_100records.csv')
```

```python
df = df.drop(['Unnamed: 0'], axis = 1) # remove first column that is the row number
```

```python
list(df)
```

```python
%matplotlib inline

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,3))
sns.distplot(df['sqftmain'], ax=axs[0])
sns.distplot(df['roll_totalvalue'], ax=axs[1])

axs[0].set_xlabel('size')
axs[1].set_xlabel('assessor value')
```

```python
sns.lmplot(x="sqftmain", y="roll_totalvalue", hue="bedrooms", col="bedrooms", data=df, aspect=.8, x_jitter=.1)
```
